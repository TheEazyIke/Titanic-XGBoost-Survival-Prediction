```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
df_train_main = pd.read_csv(r"C:\Users\iike\Documents\Projects\Personal\Kaggle\Titanic\train.csv")
df_test_main = pd.read_csv(r"C:\Users\iike\Documents\Projects\Personal\Kaggle\Titanic\test.csv")
df_test_copy = df_test_main.copy()
df_train_copy = df_train_main.copy()
df_train_main.head()
```

```python
#turn sex column into dummy variable

for i in range(len(df_train_main['Sex'])):
    if df_train_main['Sex'][i] == 'male':
        df_train_main.Sex.at[i] = 1
    else:
        df_train_main.Sex.at[i] = 0


for i in range(len(df_test_main['Sex'])):
    if df_test_main['Sex'][i] == 'male':
        df_test_main.Sex.at[i] = 1
    else:
        df_test_main.Sex.at[i] = 0
```

```python
df_train_main.Cabin.unique()
```

```python
#creates new column w/ floor level

df_train_main['Floor_Level'] = df_train_main.Cabin.str[0:1]
df_test_main['Floor_Level'] = df_test_main.Cabin.str[0:1]
```

```python
df_train_main.Floor_Level.unique()
```

```python
df_train_main.drop(columns=['Name','Ticket','Cabin','PassengerId'], inplace= True)
df_test_main.drop(columns=['Name','Ticket','Cabin','PassengerId'], inplace= True)
```

```python
#onehot encoding and imputation

cat_imputer = SimpleImputer(strategy='most_frequent')
num_imputer = SimpleImputer(strategy='median')
oh_encoder = OneHotEncoder(categories = 'auto', drop='first', handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')

#impute for train df
df_train_main_x = df_train_main.iloc[:,1:]
df_train_main_y = df_train_main.Survived

#original num imputation using column average
# age_train_null_indexes = df_train_main_x[df_train_main_x['Age'].isnull()].index.tolist()
# train_age_avg = df_train_main_x.Age.mean()
# df_train_main_x.loc[age_train_null_indexes, ['Age']] = train_age_avg


#num imputation using knn, yielded superior results
from sklearn.neighbors import KNeighborsRegressor

model_imputer = KNeighborsRegressor(weights = 'distance')
age_train_not_null_indexes = df_train_main_x[df_train_main_x['Age'].notnull()].index.tolist()
X_train_not_null = df_train_main_x[['Sex', 'Pclass', 'SibSp','Parch','Fare']].notnull()
Y_train_not_null = df_train_main_x.Age.notnull()
model_imputer.fit(X_train_not_null, Y_train_not_null)
age_train_null_indexes = df_train_main_x[df_train_main_x['Age'].isnull()].index.tolist()
X_train_null = df_train_main_x.loc[age_train_null_indexes, ['Sex', 'Pclass', 'SibSp','Parch','Fare']]
impute_predictions = model_imputer.predict(X_train_null)
df_train_main_x.loc[age_train_null_indexes, ['Age']] = impute_predictions


df_train_imputed = pd.DataFrame(cat_imputer.fit_transform(df_train_main_x))
df_train_imputed.columns = df_train_main_x.columns

#impute for test df

#original num imputation using column average
# age_test_null_indexes = df_test_main[df_test_main['Age'].isnull()].index.tolist()
# test_age_avg = df_test_main.Age.mean()
# df_test_main.loc[age_test_null_indexes, ['Age']] = test_age_avg

age_test_not_null_indexes = df_test_main[df_test_main['Age'].notnull()].index.tolist()
X_test_not_null = df_test_main[['Sex', 'Pclass', 'SibSp','Parch','Fare']].notnull()
Y_test_not_null = df_test_main.Age.notnull()
model_imputer.fit(X_test_not_null, Y_test_not_null)
age_test_null_indexes = df_test_main[df_test_main['Age'].isnull()].index.tolist()
X_test_null = df_test_main.loc[age_test_null_indexes, ['Sex', 'Pclass', 'SibSp','Parch','Fare']]
impute_predictions = model_imputer.predict(X_test_null)
df_test_main.loc[age_test_null_indexes, ['Age']] = impute_predictions

df_test_imputed = pd.DataFrame(cat_imputer.transform(df_test_main))
df_test_imputed.columns = df_test_main.columns

#onehot for train
df_train_transformed = pd.DataFrame(oh_encoder.fit_transform(df_train_imputed[['Floor_Level', 'Embarked']]))
df_test_transformed = pd.DataFrame(oh_encoder.transform(df_test_imputed[['Floor_Level', 'Embarked']]))

```

```python
df_train_transformed.isnull().sum()
```

```python
#join the transformed df's with their imputed counterparts
#make sure to drop the original column

df_train_final = df_train_imputed.join(df_train_transformed)
df_train_final.drop(columns=['Embarked', 'Floor_Level'], inplace=True)

df_test_final = df_test_imputed.join(df_test_transformed)
df_test_final.drop(columns=['Embarked', 'Floor_Level'], inplace=True)
```

```python
df_train_final.head()
```

```python
df_train_final.Pclass = df_train_final.Pclass.astype(float)
df_train_final.Sex = df_train_final.Sex.astype(float)
df_train_final.Age = df_train_final.Age.astype(float)
df_train_final.SibSp = df_train_final.SibSp.astype(float)
df_train_final.Parch = df_train_final.Parch.astype(float)
df_train_final.Fare = df_train_final.Fare.astype(float)

df_test_final.Pclass = df_test_final.Pclass.astype(float)
df_test_final.Sex = df_test_final.Sex.astype(float)
df_test_final.Age = df_test_final.Age.astype(float)
df_test_final.SibSp = df_test_final.SibSp.astype(float)
df_test_final.Parch = df_test_final.Parch.astype(float)
df_test_final.Fare = df_test_final.Fare.astype(float)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_valid, y_train, y_valid = train_test_split(df_train_final, df_train_main_y, random_state=0)

model = XGBClassifier(learning_rate=0.05, n_estimators=1000, early_stopping_rounds= 5, verbose=False )

model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)])

predictions = model.predict(X_valid)

my_model_accuracy = accuracy_score(predictions, y_valid)

print(f'The predictive accuracy is: {my_model_accuracy*100}%')
```

```python
#model hyperparameters with highest accuracy score

model_2 = XGBClassifier(learning_rate=0.03, max_depth = 7, n_estimators=1000, early_stopping_rounds= 5, verbose=False )

model_2.fit(X_train, y_train, eval_set=[(X_valid, y_valid)])

predictions_2 = model_2.predict(X_valid)

my_model_accuracy_2 = accuracy_score(predictions_2, y_valid)

print(f'The predictive accuracy is: {my_model_accuracy_2*100}%')
```

```python
model_3 = XGBClassifier(learning_rate=0.01, max_depth = 7, min_child_weight=5,  n_estimators=1000, early_stopping_rounds= 5, verbose=False )

model_3.fit(X_train, y_train, eval_set=[(X_valid, y_valid)])

predictions_3 = model_3.predict(X_valid)

my_model_accuracy_3 = accuracy_score(predictions_3, y_valid)

print(f'The predictive accuracy is: {my_model_accuracy_3*100}%')
```

```python
model_final = XGBClassifier(learning_rate=0.01, max_depth = 7, n_estimators=500, verbose=False )

model_final.fit(df_train_final, df_train_main_y)

```

```python
final_predictions = model_final.predict(df_test_final)

output = pd.DataFrame({'PassengerId': df_test_copy.PassengerId ,'Survived': final_predictions})
output.to_csv(r"C:\Users\iike\Documents\Projects\Personal\Kaggle\Titanic\titanic_submission_5.csv")
```
